```python
from simpletransformers.classification import ClassificationModel, ClassificationArgs
from google.colab import files
import pandas as pd
import io
import sklearn
```

```python
#!pip install simpletransformers
```

Loading Data and putting it in the right format

```python
uploaded = files.upload()
```

```python
OLIDA_train_df = pd.read_csv(io.BytesIO(uploaded['olid-train-small.csv']), delimiter = ',')
OLIDA_test_df = pd.read_csv(io.BytesIO(uploaded['olid-test.csv']), delimiter = ',')
Hasoc_train = pd.read_csv(io.BytesIO(uploaded['hasoc-train.csv']), delimiter = ',')
```

```python
OLIDA_test_df
```

```python
OLIDA_train_df = OLIDA_train_df.drop(['id'], axis = 1)
OLIDA_train_df.columns = ['text', 'labels']
OLIDA_train_df
```

```python
OLIDA_test_df = OLIDA_test_df.drop(['id'], axis = 1)
OLIDA_test_df.columns = ['text', 'labels']
OLIDA_test_df
```

```python
Hasoc_train_df = Hasoc_train.drop(['id'], axis = 1)
Hasoc_train_df.columns = ['text', 'labels']
Hasoc_train_df
```

BERT In Domain

```python
#Using default settings
model_args = ClassificationArgs()

model = ClassificationModel('bert', 'bert-base-cased', args = model_args)
model.train_model(OLIDA_train_df, output_dir='./bert_inDomain')
```

```python
result, model_outputs, wrong_predictions = model.eval_model(OLIDA_test_df, acc=sklearn.metrics.accuracy_score, f1 = sklearn.metrics.f1_score, recall = sklearn.metrics.recall_score, precision = sklearn.metrics.precision_score)
```

```python
result
```

```python
df_bert_wrong_inDomain = pd.DataFrame(wrong_predictions)
df_bert_wrong_inDomain.to_csv("wrong_predictions_bert_inDomain.csv")
```

BERT Cross Domain

```python
model = ClassificationModel('bert', 'bert-base-cased', args = model_args)
model.train_model(Hasoc_train_df, output_dir='./bert_crossDomain' )
```

```python
result, model_outputs, wrong_predictions = model.eval_model(OLIDA_test_df, acc=sklearn.metrics.accuracy_score, f1 = sklearn.metrics.f1_score, recall = sklearn.metrics.recall_score, precision = sklearn.metrics.precision_score)
```

```python
result
```

```python
df_bert_wrong_crossDomain = pd.DataFrame(wrong_predictions)
df_bert_wrong_crossDomain.to_csv("bert_crossDomain.csv")
```
